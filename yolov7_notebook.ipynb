## Now some changes for our custome data


### Creating a custom_data.yaml

- Go to ./data/ ---> Create a copy of coco.yaml and called: <b>custom_data.yaml </b>
- Then comment or delete all the code and paste this:

```t

train: /data/train  #
val: /data/val  # 

# number of classes
nc: 1 #Only for flowers now

# class names
names: [ 'flower' ]

```

### Creating a yolov7-custome.yaml

- Go to ./cfg/training/ ---> Create a copy of yolov7.yaml and called: <b>yolov7-custome.yaml </b>
- Then just change the <b>nc</b> variable for 1:

```t
#parameters
nc: 1 #Because now we are using just one class
...

```


## Now it is time for your own data

Put all the data in the directory ./data/

But remember 2 folders are needed at least, <b>train</b> and <b>val</b>.

And inside every folder (train and val) should be also 2 other folders, one with the <b>images</b> and other with the <b>labels</b>

If you only have the images then you can easelly label the data with https://www.makesense.ai/

Enjoy!
